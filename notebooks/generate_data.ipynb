```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
books = []

categories = [
    "Programming",
    "AI",
    "Data Science",
    "Database",
    "Analytics",
    "Mathematics"
]

for i in range(1, 101):

    books.append([
        f"B{i:03}",
        f"Book_{i}",
        f"Author_{i}",
        random.choice(categories),
        random.randint(300, 1500),
        random.randint(10, 100)
    ])

books_df = pd.DataFrame(
    books,
    columns=[
        "Book_ID",
        "Title",
        "Author",
        "Category",
        "Price",
        "Stock"
    ]
)

books_df.to_csv(
    "../data/books.csv",
    index=False
)

books_df.head()
```

```python
cities = [
    "Hyderabad",
    "Delhi",
    "Mumbai",
    "Pune",
    "Bangalore",
    "Chennai"
]

customers = []

for i in range(1, 201):

    customers.append([
        f"C{i:03}",
        f"Customer_{i}",
        random.randint(18, 60),
        random.choice(["M", "F"]),
        random.choice(cities)
    ])

customers_df = pd.DataFrame(
    customers,
    columns=[
        "Customer_ID",
        "Name",
        "Age",
        "Gender",
        "City"
    ]
)

customers_df.to_csv(
    "../data/customers.csv",
    index=False
)

customers_df.head()
```

```python
sales = []

start_date = datetime(2024, 1, 1)

for i in range(1, 5001):

    sales.append([
        f"S{i:05}",
        f"B{random.randint(1,100):03}",
        f"C{random.randint(1,200):03}",
        random.randint(1,5),
        (
            start_date +
            timedelta(
                days=random.randint(
                    0,
                    730
                )
            )
        ).strftime("%Y-%m-%d")
    ])

sales_df = pd.DataFrame(
    sales,
    columns=[
        "Sale_ID",
        "Book_ID",
        "Customer_ID",
        "Quantity",
        "Date"
    ]
)

sales_df.to_csv(
    "../data/sales.csv",
    index=False
)

sales_df.head()
```

```python
print("Books:", len(books_df))
print("Customers:", len(customers_df))
print("Sales:", len(sales_df))
```
