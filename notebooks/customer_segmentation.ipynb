```python
import pandas as pd

sales = pd.read_csv("../data/sales.csv")
customers = pd.read_csv("../data/customers.csv")
books = pd.read_csv("../data/books.csv")
```

```python
merged = pd.merge(
    sales,
    books,
    on="Book_ID"
)

merged["Revenue"] = (
    merged["Quantity"] *
    merged["Price"]
)
```

```python
customer_data = (
    merged.groupby(
        "Customer_ID"
    )["Revenue"]
    .sum()
    .reset_index()
)

customer_data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
X = customer_data[["Revenue"]]

kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

customer_data["Cluster"] = (
    kmeans.fit_predict(X)
)
```

```python
customer_data.head()
```

```python
customer_data.to_csv(
    "../reports/customer_segments.csv",
    index=False
)
```

```python
import pandas as pd
from sklearn.cluster import KMeans

sales = pd.read_csv("../data/sales.csv")
books = pd.read_csv("../data/books.csv")

merged = pd.merge(
    sales,
    books,
    on="Book_ID"
)

merged["Revenue"] = (
    merged["Quantity"] *
    merged["Price"]
)

customer_data = (
    merged.groupby("Customer_ID")
    ["Revenue"]
    .sum()
    .reset_index()
)

X = customer_data[["Revenue"]]

kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

customer_data["Cluster"] = (
    kmeans.fit_predict(X)
)

customer_data.to_csv(
    "../reports/customer_segments.csv",
    index=False
)

print("customer_segments.csv created successfully")
```
